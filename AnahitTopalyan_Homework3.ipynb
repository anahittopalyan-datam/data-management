#Queries

```python
#Groping Movies by their Rating
query = '''
SELECT Title, Rating AS Rating FROM IMDB
GROUP BY Rating
'''
pd.read_sql(query, conn)
```

```python
#Number of movies in each rating
query = '''
SELECT Rating, COUNT(Movie_id) AS movie_number FROM IMDB
GROUP BY Rating
'''
pd.read_sql(query, conn)
```

```python
#WHERE Rating > 8.5
query = '''
SELECT * FROM IMDB
WHERE Rating > 8.5
'''
pd.read_sql(query, conn)
```

```python
#WHERE Rating < 8
query = '''
SELECT * FROM IMDB
WHERE Rating < 8
'''
pd.read_sql(query, conn)
```

```python
#WHERE Total Votes > 50000
query = '''
SELECT Title, Budget, TotalVotes FROM IMDB
WHERE Budget > 18000000.0 AND TotalVotes > 50000

'''
pd.read_sql(query, conn)
```

```python
#Average Budget value rounded by 2
query = '''
SELECT ROUND(AVG(IMDB.Budget),2) AS budget_avg
FROM IMDB
'''
pd.read_sql(query, conn)
```

```python
#To understand what the Earning includes
query = '''
SELECT * FROM Earning
'''
pd.read_sql(query, conn)
```

```python
#To understand what the genre includes
query = '''
SELECT * FROM Genre
'''
pd.read_sql(query, conn)
```

```python
# Grouping movies by genre
query = '''
SELECT genre, COUNT(Movie_id) AS movie_count FROM genre
GROUP BY genre
ORDER BY movie_count DESC
'''
pd.read_sql(query, conn)
```

```python
# Top 5 Movies by Rating
query = '''
SELECT Title, Rating FROM IMDB
ORDER BY Rating DESC
LIMIT 5
'''
pd.read_sql(query, conn)
```

```python
# Top 5 Movies by Rating with Earning and Genre Information
query = '''
SELECT
    IMDB.Title,
    IMDB.Rating,
    earning.Worldwide,
    genre.genre
FROM IMDB
JOIN earning ON IMDB.Movie_id = earning.Movie_id
JOIN genre ON IMDB.Movie_id = genre.Movie_id
ORDER BY IMDB.Rating DESC
LIMIT 5
'''

pd.read_sql(query, conn)
```

```python
# Top 10 Movies by Rating with MetaCritic and Runtime
query = '''
SELECT
    IMDB.Title,
    IMDB.Rating,
    IMDB.MetaCritic,
    IMDB.Runtime
FROM IMDB
ORDER BY IMDB.Rating DESC
LIMIT 10
'''
pd.read_sql(query, conn)
```

```python
# Top 10 Movies by Rating with MetaCritic and Runtime
query = '''
SELECT genre.genre, AVG(IMDB.Rating) AS avg_rating
FROM IMDB
JOIN genre ON IMDB.Movie_id = genre.Movie_id
GROUP BY genre.genre
ORDER BY avg_rating DESC
'''
pd.read_sql(query, conn)
```

```python
# Movie Genres with Highest Earnings
query = '''
SELECT genre.genre,
       MIN(earning.Worldwide) AS min_earning,
       MAX(earning.Worldwide) AS max_earning,
       (MAX(earning.Worldwide) - MIN(earning.Worldwide)) AS spread
FROM earning
JOIN genre ON earning.Movie_id = genre.Movie_id
GROUP BY genre.genre
ORDER BY spread DESC
'''
pd.read_sql(query, conn)
```

```python
# Average Rating by Male and Female Viewers for Top Movies
query = '''
SELECT
    Title,
    Rating,
    CAST(NULLIF(VotesM, '') AS REAL) AS MaleRating,
    CAST(NULLIF(VotesF, '') AS REAL) AS FemaleRating
FROM IMDB
WHERE MaleRating IS NOT NULL AND FemaleRating IS NOT NULL
ORDER BY Rating DESC
LIMIT 10
'''
pd.read_sql(query, conn)
```

```python
# Domestic Earnings with Movie Title and Rating
query = '''
SELECT
    IMDB.Title,
    IMDB.Rating,
    earning.Domestic
FROM IMDB
JOIN earning ON IMDB.Movie_id = earning.Movie_id
ORDER BY earning.Domestic DESC
LIMIT 10
'''
pd.read_sql(query, conn)
```

```python
# Movies Grouped by MetaCritic Score, with Average Total Votes and Titles
query = '''
SELECT
    MetaCritic,
    COUNT(Movie_id) AS NumberOfMovies,
    ROUND(AVG(TotalVotes), 2) AS AverageTotalVotes,
    GROUP_CONCAT(Title, '; ') AS MoviesWithSameMetaCritic
FROM IMDB
WHERE MetaCritic IS NOT NULL AND MetaCritic != ''
GROUP BY MetaCritic
ORDER BY MetaCritic DESC
'''
pd.read_sql(query, conn)
```
